```python
# Decide on action based on missing values
# For numerical columns, fill with mean or median
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    if null_counts[col] > 0:
        df[col].fillna(df[col].median(), inplace=True)

# For categorical columns, fill with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if null_counts[col] > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv('cleaned_data_final.csv')

# Define columns to check for outliers
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# IQR method to detect and remove outliers
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Save the cleaned data after outlier removal
df.to_csv('cleaned_data_no_outliers.csv', index=False)
```

```python
import pandas as pd
from sklearn.feature_extraction import FeatureHasher

# Load your dataset
df = pd.read_csv('cleaned_data_final_no_duplicates.csv')

# Define a function to apply feature hashing
def hash_features(df, column_name, n_features=10):
    # Convert the column to a list of lists, where each list contains a single string
    data_to_hash = df[column_name].astype(str).apply(lambda x: [x])
    hasher = FeatureHasher(n_features=n_features, input_type='string')
    hashed_features = hasher.transform(data_to_hash)
    hashed_df = pd.DataFrame(hashed_features.toarray(), 
                             columns=[f'{column_name}_feature_{i}' for i in range(n_features)])
    return hashed_df

# Apply feature hashing
hashed_primary_type = hash_features(df, 'Primary Type')
hashed_location_desc = hash_features(df, 'Location Description')

# Combine with original dataframe
df = df.drop(columns=['Primary Type', 'Location Description'])
df = pd.concat([df, hashed_primary_type, hashed_location_desc], axis=1)

# Handle date column
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek
df = df.drop(columns=['Date'])

# Save the final cleaned data
df.to_csv('final_cleaned_data.csv', index=False)
```
